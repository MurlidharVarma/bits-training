# Excercise 1: Take input of number and print square and cube of that number

```python
num = input("Enter a number:")
num = eval(num)
if type(num) == int or type(num) == float:
    print("Square of the number is",num**2)
    print("Cube of the number is",num**3)
else:
    print("Please enter a valid number")
```

# Excercise 2: Take 3 numbers and print sum and average of those numbers

```python
numString = input("Enter numbers separated by space:")
numList = numString.split(" ")
total = 0
avg = 0
for n in numList:
    n = eval(n)
    if type(n) != int and type(n) != float:
        print("Please enter valid numbers")
        exit()
    else:
        total += n

print("Sum of the numbers",numString,"is",total)
print("Average of the numbers",numString,"is",total/len(numList))
```

# Excercise 3: Create Salary Statement. Ask basic salary, then calculate HRA as 10% of basic salary, add conveyance allowance of 5000 and print all the salary components and gross salary.

```python
sal = input("Enter basic salary:")
sal = eval(sal)
if type(sal)!= int and type(sal)!=float:
    print("Invalid salary")
    exit()
else:
    hra = 0.1*sal
    conveyance = 5000
    gross = sal + hra + conveyance
    print("=======================")
    print("===Salary Statement===")
    print("=======================")
    print("       Basic:",sal)
    print("         HRA:",hra)
    print("  Conveyance:",conveyance)
    print("------------------------")
    print("Gross Salary:",gross)
    print("========================")
```

# Excercise 4: Take input Principle amount, rate of interest and time period from user and calculate simple interest using formula SI = (P*R*T)/100 and total amount using formula A = P + SI

```python
P = eval(input("Enter Principle amount:"))
T = eval(input("Enter time period in years:"))
R = eval(input("Enter rate of interest:"))
SI = (P*T*R)/100
Total = P + SI
print("Simple Interest is",SI)
print("Total amount is",Total)
```

# Excercise 6: If a four digit number is entered through keyboard by user, write a program to calculate the sum of digits in the number.

```python
num = eval(input("Enter a 4 digit number:"))
n1 = num//1000
n2 = (num//100)%10
n3 = (num//10)%10
n4 = num%10
print("Sum of digits in the number",num,"is",n1+n2+n3+n4)
```

```python
num = eval(input("Enter a number:"))
sum = 0
while num <= 0:
    sum += num%10
    num = num/10

print("Sum of digits in the number is",sum)
```
